```python
# Import basic libraries
import pandas as pd
import numpy as np
# librer√≠as
import snscrape.modules.twitter as sntwitter
# Load viz library Altair
import altair as alt
#alt.renderers.enable('mimetype') #<-Este nos sirve para ver las gr√°ficas en GitHub
alt.renderers.enable('default') #<-es el m√°s com√∫n pero no refleja las gr√°ficas en GitHub
```

```python
%%time

# Par√°metros
tweets_list_ine = []
maxTweets_ine = 10_000
date_initial = "2022-11-01"

# Get tweets
for i,tweet in enumerate(sntwitter.TwitterSearchScraper('#M√©xicoDeMiVida').get_items()): # se puede a√±adir esto --> since:'+date_initial
        if i>maxTweets_ine-1:
            break
        tweets_list_ine.append([tweet.user.username, tweet.date, tweet.id, tweet.content, tweet.url, tweet.lang,
                    tweet.hashtags, tweet.likeCount, tweet.replyCount, tweet.retweetCount, tweet.quoteCount])
        # Otra info que puede sernos √∫til: tweet.media,  tweet.url
```

```python
# Pandas dataframe con tweets de los tres presidentes relacionados con el Covid-19
column_names = ("username","date","id","content","url","language","hashtags",
                "likes_count","reply_count","retweet_count","quote_count")
```

```python
# Pandas dataframe con tweets que mencionen el hashtag #INENoSeToca
df_ine = pd.DataFrame(tweets_list_ine, columns=column_names)
df_ine['date'] = df_ine['date'].dt.strftime('%Y-%m-%d')
df_ine
```

```python
source = pd.DataFrame(df_ine['date'].value_counts()).reset_index().rename(columns={'index':'fecha', 'date':"tweets"})
source
```

```python
mxn_tw = df_ine.groupby(['username','date']).agg(count=('id','count')).reset_index().sort_values(by='date')
mxn_tw
```

```python
# Gr√°fica autodescriptiva
alt.Chart(mxn_tw[mxn_tw['username']=='miseleccionmx']).mark_line().encode(
    alt.X("date",title="Fecha"),
    alt.Y('count', title="Frecuencia")
).properties(
    width=800,height=300,
    title="Tweets realizados por el usuario miseleccionmx con el hashtag MexicoDeMiVida"
)
```

```python
mxn_ln = df_ine.groupby(['language']).agg(count=('id','count')).reset_index().sort_values(by='count',ascending=False).head(10)
mxn_ln
```

```python
alt.Chart(mxn_ln).mark_arc().encode(
    theta=alt.Theta(field="count", type="quantitative"),
    color=alt.Color(field="language", type="nominal"),
).properties(
    title="Idioma de los Tweets #MexicoDeMiVida"
)
```
